```python
import requests, os
#read the predicted image file
# Make directory if it doesn't already exist
folder_name = 'image_predicted'
if not os.path.exists(folder_name):
    os.makedirs(folder_name)
    
image_predicted_url  = ["https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv"]

for url in image_predicted_url:
    response = requests.get(url)
    with open(os.path.join(folder_name, url.split('/')[-1]), mode = 'wb') as file:
        file.write(response.content)
        
df_image_predicted = pd.read_csv('image_predicted/image-predictions.tsv', sep='\t')
```

```python
#read tweet from archived tweets
df_archived_tweets = pd.read_csv('twitter-archive-enhanced.csv')
```

```python
#ref: https://www.karambelkar.info/2015/01/how-to-use-twitters-search-rest-api-most-effectively./

import tweepy
import jsonpickle

consumer_key = '' # yeYMN5z7iF91SahJ2no08wvWA
consumer_secret = ''# 12i5mFLAVBPi3W7zG0vAGoZC3N65NsOc6n28zD3DCECBNK0wwv

access_token = '' # 3296005574-9SEv7r9yh4k9NJ0wP4iB052USAnANuLjuSJd7mb
access_token_secret = '' # gld0j1MiekNgRVGhmkRX3qRMFWfWzPkNeGXaFANzsnjwF

auth = tweepy.AppAuthHandler(consumer_key, consumer_secret)

#auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth, wait_on_rate_limit = True, wait_on_rate_limit_notify = True)

if (not api):
    print ("Can't Authenticate")

tweet_id_list = df_archived_tweets.tweet_id.tolist()
tweet_is_list_error = []

print("Downloading tweets ")

with open('tweet_json.txt', 'w') as outfile:
    for tweet_id in tweet_id_list:
        try:
            tweet = api.get_status(tweet_id, tweet_mode='extended')
            outfile.write(jsonpickle.encode(tweet._json, unpicklable=False) + '\n')
        except tweepy.TweepError as e:
            tweet_is_list_error.append(tweet_id)

print(tweet_is_list_error.count)
print ("Downloaded tweets")
```

```python
#reading the tweet_json.txt and creating a data frame from it
import json

df_tweet_json = []

with open('tweet_json.txt', 'r') as json_file:  
    for line in json_file:
        j_content = json.loads(line)
        df_tweet_json.append({'id': j_content['id'], 
                              'favorite_count': j_content['favorite_count'],
                              'retweet_count': j_content['retweet_count']})
        
df_tweet_json = pd.DataFrame(df_tweet_json)
```
